```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create data
x = np.linspace(0, 10, 50)
df = pd.DataFrame({
    'x': x,
    'y1': x,           # y = x
    'y2': 0.5 * x,     # y = 0.5x
    'y3': np.sin(x),   # Example
    'y4': np.cos(x),   # Example
    'y5': np.exp(0.1*x),  # Example
    'y6': np.log(x+1)     # Example
})

# Create 3x2 subplot grid
fig, axs = plt.subplots(3, 2, figsize=(10, 10))

# Plot 1: y = x
axs[0, 0].plot(df['x'], df['y1'], label='y = x', color='blue')
axs[0, 0].legend()

# Plot 2: y = 0.5x with point (3, 2)
axs[0, 1].plot(df['x'], df['y2'], label='y = 0.5x', color='green')
axs[0, 1].legend()

# Plot 3: Example - sine
axs[1, 0].plot(df['x'], df['y3'], label='y = sin(x)', color='orange')
axs[1, 0].legend()

# Plot 4: Example - cosine
axs[1, 1].plot(df['x'], df['y4'], label='y = cos(x)', color='purple')
axs[1, 1].legend()

# Plot 5: Example - exponential
axs[2, 0].plot(df['x'], df['y5'], label='y = exp(0.1x)', color='brown')
axs[2, 0].legend()

# Plot 6: Example - logarithmic
axs[2, 1].plot(df['x'], df['y6'], label='y = log(x+1)', color='pink')
axs[2, 1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Sample data
x = np.linspace(0, 10, 50)
y = x

# 1st plot: y = x, green squares
plt.subplot(3, 2, 1)
plt.plot(x, y, 'gs')

# 2nd plot: y = 0.5x, red squares + point at (3, 2)
plt.subplot(3, 2, 2)
plt.plot(x, 0.5*y, 'rs')

# 3rd plot: y = x, red dashed line
plt.subplot(3, 2, 3)
plt.plot(x, y**2, 'r--')

# 4th plot: y = x, red dashed line
plt.subplot(3, 2, 4)
plt.plot(x, np.sqrt(y), 'r--')

# 5th plot: y = sin(x), red dashed line
plt.subplot(3, 2, 5)
plt.plot(x, np.sin(y), 'r--')

# 6th plot: y = cos(x), red dashed line
plt.subplot(3, 2, 6)
plt.plot(x, np.cos(y), 'r--')

plt.tight_layout()
plt.show()
```
